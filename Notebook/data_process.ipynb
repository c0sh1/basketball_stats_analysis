## CONTENT

This Dataset Includes:

### 📊 Dataset Overview
- **Time Period:** Seasons 1999-2020
- **Coverage:** 49 Leagues, ~11K players with details & stats per season

### 👤 Player Details
- Birth Date
- Height & Weight
- Nationality
- High School

### 📈 Season Statistics
- Scoring Stats
- Free Throws
- Rebounds
- Blocks
- Assists
- Minutes Played
- Games Played
- And more...

### 📝 Additional Information
This dataset goes beyond simple rows and columns. It provides comprehensive player information and detailed performance metrics across multiple decades and leagues, making it accessible for analysis and research.
