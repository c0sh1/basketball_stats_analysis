## Importamos y pegamos

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Data/stats_by_season.csv')

df
```

```python
df.columns
```

#### Descartamos las columnas innecesarias y verificamos si hay algún dato nulo dentro del DF

```python
df = df.drop(columns=['height','weight','birth_month','birth_date'])
```

```python
df.isnull().any().any()
```

#### Al devolver 'True' nos dice que sí que lo hay. <br> a continuación buscamos en que columna/s están en %

```python
print((df.isnull().sum() / len(df) *100).round(2))
```

```python
df[['Team', 'nationality', 'high_school', 'draft_round', 'draft_pick', 'draft_team']].info()
```

```python
df['Team'] = df['Team'].fillna('UNK')
df[['nationality', 'high_school']] = df[['nationality', 'high_school']].fillna('Unknown')
df[['draft_round', 'draft_pick']] = df[['draft_round', 'draft_pick']].fillna(np.nan)
df['draft_team'] = df['draft_team'].fillna('Undrafted')

print((df.isnull().sum() / len(df) *100).round(2))
```

```python
df
```

```python
dff = df.copy()
```

#### Para los datos nulos que quedan: 'height_cm' 'birth_yea' 'weight_kg' usaremos KNN, <br> puesto que creo que un jugador pivot no medirá lo mismo que un base.

#### Primero estandarizamos y KNN

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import KNNImputer

cols = ["height_cm", "weight_kg", "birth_year"]

scaler = StandardScaler()
X = scaler.fit_transform(dff[cols])                 # 1. escalar (los NaN se quedan como NaN)
X = KNNImputer(n_neighbors=10).fit_transform(X)    # 2. imputar en escala z
dff[cols] = scaler.inverse_transform(X) 
```

```python
print((dff.isnull().sum() / len(dff) * 100).round(2))
```

#### Queremos ver si hay algún espacio de más en las celdas de las columnas o algún caracter de más.

```python
# Recorremos a ver si hay 

for col in dff.select_dtypes(include='object').columns:
    print(f"\n{col}:")
    print(dff[col].unique()[:10])
```

#### Vemos que hay espacios de más. Procedemos a eliminarlos: 'Bethel High School ', 'Robert G. Cole High School '

```python
cols_txt = dff.select_dtypes(include='str').columns
dff[cols_txt] = dff[cols_txt].apply(lambda x: x.str.strip())
```

```python
dff['high_school'].unique()[:10]
```

```python
dff
```

#### Para no tener problemas con los NaN del draft pick y round, creamos una nueva columna con undrafted. <br> Si está Drafteado un 1, si no lo está un 0

```python
dff['drafted'] = df.apply(
    lambda row: 0 if (pd.isna(row['draft_round']) or 
                      pd.isna(row['draft_pick'])  or 
                      row['draft_team'] == 'Undrafted') 
                else 1, axis=1
)

# Verificar
print(dff['drafted'].value_counts())
```

#### No tenemos ya ningún dato nulo. Procedemos al guardado del DF

```python
dff.to_csv('../Data/stsbyseasonclean.csv', index=False)
```
