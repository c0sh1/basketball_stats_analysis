```python
df = df.drop(columns=['height','weight','birth_month','birth_date'])
```

#### Verificamos si hay algún dato nulo dentro del DF y si los hay, saber en que columnas

```python
df.isnull().any().any()
```

#### Al devolver 'True' nos dice que sí que lo hay. <br> a continuación buscamos en que columna/s están en %

```python
print((df.isnull().sum() / len(df) *100).round(2))
```

```python
df[['Team', 'nationality', 'high_school', 'draft_round', 'draft_pick', 'draft_team']].info()
```

```python
df['Team'].value_counts()
```

```python
df['nationality'].value_counts()
```

#### Rellenamos los campos nulos:
###### En Team y nationality: Unknown <br> en draft_team: undrafted <br> en draft_round y _pick: un '0'

```python
df['Team'] = df['Team'].fillna('UNK')
df['nationality'] = df['nationality'].fillna('Unknown')
df[['draft_round', 'draft_pick']] = df[['draft_round', 'draft_pick']].fillna(0)
df['draft_team'] = df['draft_team'].fillna('Undrafted')

print((df.isnull().sum() / len(df) *100).round(2))
```

###### En high_school también colocaremos 'unknown'

```python
df['high_school'] = df['high_school'].fillna('Unknown')

print((df.isnull().sum() / len(df) *100).round(2))
```

```python
df_c = df.copy()
```

```python
df_c
```

#### Para los datos nulos que quedan: 'height_cm' 'birth_yea' 'weight_kg' usaremos KNN, <br> puesto que creo que un jugador pivot no medirá lo mismo que un base.

```python
from sklearn.impute import KNNImputer
```

```python
knn = KNNImputer(n_neighbors=20)

cols = ['height_cm', 'weight_kg', 'birth_year']

df_c[cols] = knn.fit_transform(df_c[cols])

df_c[cols].isnull().sum()
```

```python
print((df_c.isnull().sum() / len(df) *100).round(2))
```

#### Queremos ver si hay algún espacio de más en las celdas de las columnas o algún caracter de más.

```python
# Recorremos a ver si hay 

for col in df_c.select_dtypes(include='object').columns:
    print(f"\n{col}:")
    print(df_c[col].unique()[:10])
```

#### Vemos que hay espacios de más. Procedemos a eliminarlos: 'Bethel High School ', 'Robert G. Cole High School '

```python
cols_txt = df_c.select_dtypes(include='str').columns
df_c[cols_txt] = df_c[cols_txt].apply(lambda x: x.str.strip())

```
